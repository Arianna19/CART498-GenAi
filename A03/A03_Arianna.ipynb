```python
from openai import OpenAI
from google.colab import userdata
import random

# Initialize OpenAI client
client = OpenAI(api_key=userdata.get("OPENAI_API_KEY"))

print("(ﾉ◕ヮ◕)ﾉ*:･ﾟ✧ GPT-4.1, The Worst Mathematician Ever ✧ﾟ･:*ヽ(◕ヮ◕ヽ)")
print("I will attempt math. Please lower your expectations.\n")

# User input
n = int(input("Enter a base number (n): "))
i = int(input("Enter number of iterations (i): "))

current = n
successful_steps = 0
mistakes = 0
consecutive_failures = 0

print("\n(ง •̀_•́)ง Beginning calculations...\n")

# Emotional states
happy_faces = ["(≧◡≦)", "(•̀ᴗ•́)و", "(๑˃ᴗ˂)ﻭ", "(ﾉ◕ヮ◕)ﾉ"]
confused_faces = ["(⊙_⊙;)", "(・_・?)", "(・・?)"]
sad_faces = ["(ಥ﹏ಥ)", "(╥﹏╥)", "(ノ_<。)"]
broken_faces = ["(x_x)", "(＠_＠)", "(…)", "(;_;)"]

happy_lines = [
    "WAIT— I DID IT.",
    "YES. A NUMBER. A CORRECT ONE.",
    "Math fears me today.",
    "I am briefly powerful."
]

confused_lines = [
    "Hmm… that feels suspicious.",
    "I don't trust this number.",
    "Why does math look like this?"
]

sad_lines = [
    "I messed it up again.",
    "Numbers hate me.",
    "I should apologize to math."
]

broken_lines = [
    "I can't do this anymore.",
    "Everything is numbers and I am afraid.",
    "Why did you make me do math."
]

def emotion(fails):
    if fails == 0:
        return random.choice(happy_faces), random.choice(happy_lines)
    elif fails <= 2:
        return random.choice(confused_faces), random.choice(confused_lines)
    elif fails <= 4:
        return random.choice(sad_faces), random.choice(sad_lines)
    else:
        return random.choice(broken_faces), random.choice(broken_lines)

while successful_steps < i:

    step = successful_steps + 1
    correct_answer = current * current

    # Call OpenAI (legit API usage)
    response = client.responses.create(
        model="gpt-4.1-nano",
        temperature=1.4,
        top_p=0.7,
        input=(
            f"Quickly estimate without double-checking:\n"
            f"What is {current} multiplied by itself?\n"
            f"Reply with only a number."
        )
    )

    # GPT's raw output (we ignore correctness on purpose)
    try:
        gpt_base = int(response.output_text.strip())
    except:
        gpt_base = correct_answer

    # FORCE GPT TO BE WRONG MOST OF THE TIME
    if random.random() < 0.7:  # 70% chance to hallucinate
        error_factor = random.choice([-3, -2, -1, 1, 2, 3])
        gpt_answer = correct_answer + error_factor
    else:
        gpt_answer = correct_answer

    print(f"{emotion(consecutive_failures)[0]} Iteration {step}")
    print(f"GPT-4.1 says: {gpt_answer}")
    print(f"Correct answer: {correct_answer}")

    if gpt_answer == correct_answer:
        # Success → move forward
        consecutive_failures = 0
        successful_steps += 1
        current = correct_answer
        face, line = emotion(0)
        print(face, line)
    else:
        # Failure → retry same iteration
        mistakes += 1
        consecutive_failures += 1
        face, line = emotion(consecutive_failures)
        print(face, line)

    print("-" * 55)

print("\n(… ) Experiment complete.")
print(f"Total mistakes made: {mistakes}")

if mistakes <= 2:
    print("Final diagnosis: suspicious competence.")
elif mistakes <= 5:
    print("Final diagnosis: moderate numerical distress.")
elif mistakes <= 8:
    print("Final diagnosis: severe mathematical despair.")
else:
    print("Final diagnosis: total cognitive collapse.")
```
